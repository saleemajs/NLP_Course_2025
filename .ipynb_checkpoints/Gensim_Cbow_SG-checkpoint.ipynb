```python
from keras.models import Model
from keras.layers import Input, Embedding, Dense, Lambda
import keras.backend as K

# Parameters
vocab_size = 5000  # Vocabulary size
embedding_dim = 100  # Embedding dimensions
context_size = 4  # Number of context words

# Input layer
context_words = Input(shape=(context_size,))

# Embedding layer
embedding = Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=context_size)(context_words)

# Average embeddings of context words
average_embedding = Lambda(lambda x: K.mean(x, axis=1))(embedding)

# Dense layer to predict target word
output = Dense(vocab_size, activation='softmax')(average_embedding)

# Model
cbow_model = Model(inputs=context_words, outputs=output)
cbow_model.compile(optimizer='adam', loss='categorical_crossentropy')
cbow_model.summary()

```

```python
#2. Skip-gram Implementation

from keras.layers import Dot, Reshape
from keras.models import Model

# Input layers
target_word = Input(shape=(1,))
context_word = Input(shape=(1,))

# Embedding layer (shared for both target and context words)
embedding = Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=1)

target_embedding = embedding(target_word)
context_embedding = embedding(context_word)

# Dot product of target and context embeddings
dot_product = Dot(axes=-1)([target_embedding, context_embedding])
output = Reshape((1,))(dot_product)

# Model
skipgram_model = Model(inputs=[target_word, context_word], outputs=output)
skipgram_model.compile(optimizer='adam', loss='binary_crossentropy')
skipgram_model.summary()
```

```python
from gensim.models import Word2Vec

# Example corpus
corpus = [['I', 'love', 'eating', 'pizza'],
          ['Pizza', 'is', 'my', 'favorite', 'food'],
          ['I', 'enjoy', 'eating', 'burgers']]

# CBOW Model
cbow_model = Word2Vec(sentences=corpus, vector_size=100, window=5, sg=0)

# Skip-gram Model
skipgram_model = Word2Vec(sentences=corpus, vector_size=100, window=5, sg=1)

# Access word embeddings
word = 'pizza'
cbow_embedding = cbow_model.wv[word]
skipgram_embedding = skipgram_model.wv[word]
```
